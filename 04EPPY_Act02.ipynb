```python
def contarNaN(df):
    # Contamos la cantidad de valores NaN en cada columna
    cantidad_nan_por_columna = df.isna().sum()
    
    # Contamos la cantidad total de valores NaN 
    cantidad_total_nan = df.isna().sum().sum()
    
    # Imprimimos la cantidad de valores NaN por columna
    print("Cantidad de NaN por columna:")
    print(cantidad_nan_por_columna)
    
    # Imprimimos la cantidad total de valores NaN
    print("Cantidad total de NaN en el DataFrame:", cantidad_total_nan)
```

### 2.1 Obtener un gráfico de barras con la evolución del número de deportes por año, eje X años, eje Y número de deportes

```python
import matplotlib.pyplot as plt
 
# Agrupamos por año y contamos el número de deportes únicos por año con nunique
conteo_deportes_por_año = df_athlete_data_filtered.groupby('Year')['Sport'].nunique()

# Creamos el gráfico de barras con Matplotlib
plt.figure(figsize=(12, 6))  # Tamaño del gráfico
bars = plt.bar(conteo_deportes_por_año.index, conteo_deportes_por_año.values, color='skyblue')
plt.xlabel('Año')
plt.ylabel('Número de Deportes')
plt.title('Evolución del Número de Deportes por Año')
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7)  

# Agregamos el año y la letra vertical sobre cada barra
for bar, año, deportes in zip(bars, conteo_deportes_por_año.index, conteo_deportes_por_año.values):
    plt.text(bar.get_x() + bar.get_width() / 2, deportes, str(año), ha='center', va='bottom', rotation='vertical')

plt.tight_layout()

# Mostramos el grafico en barras
plt.show()
```

En la grafica de barras se observa que hasta el año 1992 la cantidad de deportes ha ido en credimiento,y luego entre el año 1994 y 2016 se ha estancado entre 10 y 35 deportes

### 2.2. Gráfico con la evolución de la participación femenina en las olimpiadas a lo largo de la historia, eje X años, eje Y número de atletas, para cada año una barra para cada sexo

```python
import matplotlib.pyplot as plt

df=df_athlete_data_filtered

# Agrupamos los datos por año y sexo y contamos el número de atletas de cada sexo en cada año
grouped = df.groupby(['Year', 'Sex']).size().unstack()

# Creamos un gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(12, 6))

# Obtenemos los años como etiquetas en el eje X
years = grouped.index

# Obtenemos los datos de participación masculina y femenina para cada año
participacion_masculina = grouped['M']
participacion_femenina = grouped['F']

# Establecemos el ancho de las barras
bar_width = 0.35

# Establecemos las Coordenadas para las barras de hombres y mujeres
x_male = range(len(years))
x_female = [x + bar_width for x in x_male]

# Creamos las barras
ax.bar(x_male, participacion_masculina, width=bar_width, label='Masculino', color='blue')
ax.bar(x_female, participacion_femenina, width=bar_width, label='Femenino', color='red')

# Configuramos etiquetas y título
ax.set_xlabel('Año')
ax.set_ylabel('Número de Atletas')
ax.set_title('Evolución de la Participación en las Olimpiadas por Género')

# Mostramos la leyenda
ax.legend()

# Configuramos las etiquetas del eje X
ax.set_xticks([x + bar_width / 2 for x in x_male])
ax.set_xticklabels(years)

# Rotamos las etiquetas del eje X para que sean legibles
plt.xticks(rotation=45)

# Mostramos el gráfico
plt.tight_layout()
plt.show()
```

En la grafica de barras obseramos que a partir del año 1924  ha empezado a incrementarse la evolucion de participacion del grupo femenino en las Olimpiadas, a diferencia del grupo masculino que inicio desde 1896 hasta tener un pico mas alto en el año 1992, por lo que desde el año 1992 se ha estabilizado tanto el grupo masculino como femenino, asimismo de todo la grafica se afirma que el grupo masculino tiene un mayor impacto ya que abarcan un numero mayor de atletas en comparacion al grupo femenino hasta la fecha del año 2016

### Paso 3: Normalizacion de Segundos (Patron: d.d)


Debido a que lo que se quiere predecir es el tiempo para una carrera de 100m lisos, lo correcto seria filtrar valores que tengan formato en segundos, ya que los : (dos puntos) es empleado principalmente para representar minutos y horas.¿, y estos seran eliminados y solo filtrados los que coincidan con el patron

```python
# Utilizamos str.contains() para filtrar las filas con el patrón decimal.decimal
pattern = r'\d{2}+\.\d{1,3}+' 
df_csv_results_sel_sinnan = df_csv_results_sel_sinnan[df_csv_results_sel_sinnan['Result'].str.match(pattern)]

display(df_csv_results_sel_sinnan)
#Mostramos los resultados unicos en Result, para verificar que el filtro este correcto
print(df_csv_results_sel_sinnan['Result'].unique().tolist())
```

Convertimos la columna Result en valor numerico

```python
import pandas as pd 
# Convierte la columna 'Result_Minutos_P2' a tipo float, estableciendo valores no convertibles como NaN
df_csv_results_sel_sinnan['Result'] = pd.to_numeric(df_csv_results_sel_sinnan['Result'], errors='coerce')

# Imprime el DataFrame con la columna convertida a float
display(df_csv_results_sel_sinnan)
```

## MACHINE LEARNING - SVR

Entrenamos la data mediante la tecnica de machine learning llamada SVR

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
 

# Dividir los datos en características (X) y etiquetas (y)
X = df_resultante[['Sex_M', 'Age', 'Height', 'Weight']]
y = df_resultante['Result']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=15)

# Normalizar las características para mejorar el rendimiento del SVM
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Crear un modelo de SVM para regresión (SVR)
svr_model = SVR(kernel='linear')  # Puedes ajustar el kernel según tus necesidades, aquí se usa un kernel lineal

# Entrenar el modelo
svr_model.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = svr_model.predict(X_test)

# Evaluar el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE (Error Cuadrático Medio): {mse}')
print(f'R2 Score (Coeficiente de Determinación): {r2}')

```

## MACHINE LEARNING - RANDOM FOREST

Entrenamos la data mediante la tecnica de machine learning llamada Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Cargamos el DataFrame
data = df_resultante 

# Dividimos los datos en conjuntos de entrenamiento y prueba
X = data[['Sex_M', 'Age', 'Height', 'Weight']]
y = data['Result']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)

# Creamos y entrenamos un modelo Random Forest
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)  
rf_regressor.fit(X_train, y_train)

# Realizamos predicciones en el conjunto de prueba
y_pred = rf_regressor.predict(X_test)

# Evaluamos el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse}')
print(f'R2 Score: {r2}')
```

## MACHINE LEARNING - ARBOL DE DECISIONES

Entrenamos la data mediante la tecnica de machine learning llamada Arbol de decisiones

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Dividimos los datos en características (X) y etiquetas (y)
X = df_resultante[['Sex_M', 'Age', 'Height', 'Weight']]
y = df_resultante['Result']

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creamos un modelo de árbol de decisiones para regresión
model_tree = DecisionTreeRegressor(random_state=42)

# Entrenamos el modelo
model_tree.fit(X_train, y_train)

# Realizamos predicciones en el conjunto de prueba
y_pred = model_tree.predict(X_test)

# Evaluamos el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE (Error Cuadrático Medio): {mse}')
print(f'R2 Score (Coeficiente de Determinación): {r2}')
```
